```python
import random
import numpy as np
import matplotlib.pyplot as plt

num = 100000
seed_arr = np.random.rand(num) * 2 * np.pi

arr_sin = np.sin(seed_arr)*141
hist = np.histogram(arr_sin, bins=1000)[0] /num

hy, bins = np.histogram(arr_sin, bins = 100)
hy = hy/num
hx = bins[:-1] + np.diff(bins)/2
hx_ = sorted(arr_sin)
hy_ = [i/num for i in range(num)]

plt.plot(hx,hy)
plt.title('PDF of $141sinx$')
plt.show()

plt.plot(hx_, hy_)
plt.title('CDF of $141sinx$')
plt.show()
```

## 説明

6行目で y = 141sin(x) の 'x' に相当する配列を作成

8行目で 'y' に相当する配列を作成

11行目で 'y' の度数の一覧の配列とビンの境界の配列のタプルを取得
