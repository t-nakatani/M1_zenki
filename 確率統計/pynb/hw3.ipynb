```python
# import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 1000000
result = []
for i in range(N):
    arr = np.random.random_sample(5) * 2 - 1 #[0,1] -> [-1, 1]
    cumsum = np.cumsum(arr)
    result.append(cumsum)
df = pd.DataFrame(result)
```

```python
n_bins = 50
for i in range(5):
    hy, bins = np.histogram(df[i], bins = n_bins*(i+1))
    hy = hy/N*n_bins*(i+1)/(bins[-1]-bins[0])
    hx = bins[:-1] + np.diff(bins)/2
    plt.plot(hx,hy)
plt.show()
```
