```python
import numpy as np
```

```python
D = 2
DL = D + 1
N = 2
L = 0.1
```

```python
# input data: Label, input1, input2 
data = [1, 0, 1, 0, 1, 0]
def sigm(x):
    return 1.0 / (1 + np.exp(-x))
```

```python
loss1 = []
w = [0] * DL
for j in range(1000):
    dw1 = [0] * DL
    e = 0
    for i in range(N):
        
        w1 = sum([d * w for d, w in zip(data[DL*i+1:DL*i+2], w[:-1])]) + w[D]
        w1 = sigm(w1)
        e += (data[i*DL] - w1) ** 2
        dw1[0] += - (data[i*DL] - w1) * (1 - w1) * w1 * data[i*DL+1]
        dw1[1] += - (data[i*DL] - w1) * (1 - w1) * w1 * data[i*DL+2]
        dw1[2] += - (data[i*DL] - w1) * (1 - w1) * w1 * 1
    loss1.append(e)
    w = [w_ - dw_*L for w_, dw_ in zip(w, dw1)]



loss2 = []
w = [0] * DL
for j in range(1000):
    dw1 = [0] * DL
    e = 0
    for i in range(N):
        
        w1 = sum([d * w for d, w in zip(data[DL*i+1:DL*i+2], w[:-1])]) + w[D]
        w1 = sigm(w1)
        e += (data[i*DL] - w1) ** 2
        dw1[0] += - (data[i*DL] - w1) * w1 * data[i*DL+1]
        dw1[1] += - (data[i*DL] - w1) * w1 * data[i*DL+2]
        dw1[2] += - (data[i*DL] - w1) * w1 * 1
    loss2.append(e)
    w = [w_ - dw_*L for w_, dw_ in zip(w, dw1)]
```

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(1000), loss1, label='loss1')
plt.plot(np.arange(1000), loss2, label='loss2')
plt.legend()
plt.show()
```
