## BGA22078 中谷友哉

この課題では，簡単化のために回帰関数の推定値のハット記号を省略する．  
実数関数$f(x, y)$と定数$c$を用いた$g(x) = f(x, c)$を考えると
$$
\min _{x,y}f(x, y) \leq \min_{x}g(x) = \min_{x\in R, y=c}f(x, y)
$$
が成り立つ．
これは$x$がベクトル$\boldsymbol{x}$の時も同様であるから，
$$\boldsymbol{\beta}=(\beta_0, ..., \beta_d)$$
$$
\min_{\boldsymbol{\beta}, \beta_{d+1}}f(\boldsymbol{\beta}, \beta_{d+1}) = MSE(d+1)
$$

$$
\min_{\boldsymbol{\beta}}g(\boldsymbol{\beta}) = \min_{\boldsymbol{\beta}, \beta_{d+1} =0}f(\boldsymbol{\beta}, \beta_{d+1})=MSE(d)
$$
とすると，
$$
MSE(d+1) = \min_{\boldsymbol{\beta}, \beta_{d+1}}f(\boldsymbol{\beta}, \beta_{d+1}) \leq \min_{\boldsymbol{\beta}}g(\boldsymbol{\beta}) = \min_{\boldsymbol{\beta}, \beta_{d+1} =0}f(\boldsymbol{\beta}, \beta_{d+1})=MSE(d)
$$
が成り立つ．
よって$MSE(d) \geq MSE(d+1)$が示された．
