# 1-a

上の$L$を$\pi_k$で微分すると，次の通りになる．
$$
\frac{\partial L}{\partial \pi_k} = \frac{\sum_{i=1}^{n}y_{ik}}{\pi_k} - \lambda
$$
同様に，上の$L$を$\lambda$で微分すると，次の通りになる．
$$
\frac{\partial L}{\partial \lambda} = \sum_{k=1}^{K} \pi_k-1
$$

# 1-b
$$
\frac{\partial L}{\partial \pi_k} = \frac{\sum_{i=1}^{n}y_{ik}}{\pi_k} - \lambda = 0
$$
この等式を解くと次の制約条件が導かれる．
$$
\pi_k = \frac{\sum_{i=1}^{n}y_{ik}}{\lambda}
$$

# 1-c
$$
\frac{\partial L}{\partial \lambda} = \sum_{k=1}^{K} \pi_k-1
$$
1-bと同様に，この等式を解くと次の制約条件が導かれる．
$$
\sum_{k=1}^{K} \pi_k=1
$$
この制約条件の左辺は次のように変形できる．
$$
\sum_{k=1}^{K} \pi_k = 
\sum_{k=1}^{K} \frac{\sum_{i=1}^{n}y_{ik}}{\lambda} = 
\frac{1}{\lambda}\sum_{k=1}^{K} \sum_{i=1}^{n}y_{ik} = 
\frac{1}{\lambda} \times n = 
\frac{n}{\lambda}
$$
よって，$\frac{n}{\lambda}=1$から$\lambda=n$が導かれるため，MLE'sは$\lambda=n$としたラグランジュの未定乗数法により実現することができる．

$$
\sum_{i=1}^{n}y_{ik} \boldsymbol{\mu}_k = \sum_{i=1}^{n}y_{ik} \boldsymbol{x}_i 
$$
$$
\boldsymbol{\mu}_k = \frac{\sum_{i=1}^{n}y_{ik}\boldsymbol{x}_i }{\sum_{i=1}^{n}y_{ik}}
