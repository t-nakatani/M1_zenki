$n_k^{(t)}, r_{ik}^{(t)}, $の定義より次の式が成り立つ．
$$
\sum_{k=1}^{K}n_k^{(t)} = \sum_{k=1}^{K}\sum_{i=1}^{n}r_{ik}^{(t)} = \sum_{k=1}^{K}\sum_{i=1}^{n}
\frac{\pi_k^{(t)}f(\boldsymbol{x}_i |\boldsymbol{\mu}_k^{(t)}, \Sigma_k^{(t)})}
{\sum_{j=1}^K\pi_j^{(t)}f(\boldsymbol{x}_i |\boldsymbol{\mu}_j^{(t)}, \Sigma_j^{(t)})}
$$

また，summationの順番を入れ替えることで次のように変形でき，$\sum_{k=1}^K n_k^{(t)} = n$を示すことができる．　　
$$
 \sum_{k=1}^{K}\sum_{i=1}^{n}
\frac{\pi_k^{(t)}f(\boldsymbol{x}_i |\boldsymbol{\mu}_k^{(t)}, \Sigma_k^{(t)})}
{\sum_{j=1}^K\pi_j^{(t)}f(\boldsymbol{x}_i |\boldsymbol{\mu}_j^{(t)}, \Sigma_j^{(t)})}
 =

{\sum_{j=1}^K\pi_j^{(t)}f(\boldsymbol{x}_i |\boldsymbol{\mu}_j^{(t)}, \Sigma_j^{(t)})}
 = 
$$$$

{\sum_{j=1}^K\pi_j^{(t)}f(\boldsymbol{x}_i |\boldsymbol{\mu}_j^{(t)}, \Sigma_j^{(t)})}

# 3.

2.で求めたラグランジュ関数$L$を$\pi_k$で微分すると，次の通りになる．
$$
\frac{\partial L}{\partial \pi_k} = \frac{\sum_{i=1}^{n}r_{ik}^{(t)}}{\pi_k} - \lambda
$$
同様に，上の$L$を$\lambda$で微分すると，次の通りになる．
$$
\frac{\partial L}{\partial \lambda} = \sum_{k=1}^{K} \pi_k-1
$$

ここで$\frac{\partial L}{\partial \pi_k}=0$とすると，$\pi_k$は次のように求められる．
$$
\frac{\partial L}{\partial \pi_k} = \frac{\sum_{i=1}^{n}r_{ik}^{(t)}}{\pi_k} - \lambda = 0
$$

$$\pi_k = \frac{\sum_{i=1}^{n}r_{ik}^{(t)}}{\lambda}$$

同様に，$\frac{\partial L}{\partial \lambda}=0$とすると，次の条件が導かれる．

$$
\sum_{k=1}^{K} \pi_k=1
$$
この制約条件の左辺は次のように変形できる．
$$
\sum_{k=1}^{K} \pi_k = 
\sum_{k=1}^{K} \frac{\sum_{i=1}^{n}r_{ik}^{(t)}}{\lambda} = 
\frac{1}{\lambda}\sum_{k=1}^{K} \sum_{i=1}^{n}r_{ik}^{(t)} = 
\frac{1}{\lambda} \times n = 
\frac{n}{\lambda}
$$
よって，$\frac{n}{\lambda}=1$から$\lambda=n$が導かれるため，MLE'sは$\lambda=n$としたラグランジュの未定乗数法により実現することができる．
そこで，$\pi_k = \frac{\sum_{i=1}^{n}r_{ik}^{(t)}}{\lambda}$と$\lambda=n$から，
$$\pi_k^{(t+1)} = \frac{\sum_{i=1}^{n}r_{ik}^{(t)}}{\lambda} = \frac{1}{n}\sum_{i=1}^{n}r_{ik}^{(t)}$$
