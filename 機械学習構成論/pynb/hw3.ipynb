Kのknotを持つ3次スプライン関数は，knotによりK+1の領域に分割される．  
1つの領域ごとに，3次関数は4つのパラメータを持つため，パラメータの数は全ての領域で4(K+1)個になる．

ただし，各knotにおいて関数そのものの連続性と，1次・2次導関数の連続性を満たす必要があるため，knot1つにつき3つの制約式つまり全体で3Kこの制約式が生まれる．
例として三次スプライン関数$g(x)$のknot$c_k$での制約式は，$g(c_k-0) = g(c_k+0)$，$g'(c_k-0) = g'(c_k+0)$，$g''(c_k-0) = g''(c_k+0)$の3つである．  
自由度 = パラメータの数 - 制約式の数より，$自由度=4(K+1) - 3K = K + 4$となる
