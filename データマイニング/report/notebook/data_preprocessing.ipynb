```python
pth_data = '/Users/nakatani/Desktop/22//report/data/img/xxx/*.png'
pth_mask = '/Users/nakatani/Desktop/22//report/data/mask/xxx/*.png'
labels = ['AnemoneHybrida', 'AnemoneNikoensis', 'AnemonePulsatilla']
all_data = []
for label in labels:
    imgs = sorted(glob.glob(pth_data.replace('xxx', label)))
    masks = sorted(glob.glob(pth_mask.replace('xxx', label)))
    data = []
    for path_img, path_mask in zip(imgs[:15], masks[:15]):
        img = cv2.imread(path_img)
        mask = cv2.imread(path_mask, 0)
        hist3c = []
        for i in range(3):
            img_1c = img[:, :, i]
            arr = img_1c[mask==255]
            hist1c, bins = np.histogram(arr, bins=np.linspace(5, 255, 11))
            hist3c.extend(hist1c)
        all_data.append(hist3c)
#     print(np.array(data).shape)
#     all_data.append(data)
print(np.array(all_data).shape)

import random
random.seed(0)
df = pd.DataFrame(all_data)
df['label'] = [i//15 for i in range(45)]
data_idx_arr = random.sample(range(45), 10)
print(f'data_idx_arr: {sorted(data_idx_arr)}')
for data_idx in data_idx_arr:
    feature_idx = np.random.randint(30)
    df.iloc[data_idx, feature_idx] = np.nan
df.to_csv('../data/30feature_45sample_incomp.csv', index=False)
```
