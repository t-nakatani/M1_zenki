```python
from sklearn.datasets import load_iris
import numpy as np

data_iris = load_iris()
X = data_iris.data
y = data_iris.target
```

```python
len(y[y==0]), len(y[y==1]), len(y[y==2])
```

```python
print(np.sum(X[:50], 0) /50)
print(np.sum(X[50:100], 0) /50)
print(np.sum(X[100:150], 0) /50)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:50, 0], X[:50, 1], color='#87cefa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='red', marker='x')
plt.scatter(X[100:150, 0], X[100:150, 1], color='#adff2f', marker='^')
plt.show()
```
