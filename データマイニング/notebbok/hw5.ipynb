```python
import pandas as pd
import numpy as np
list2d = [[1, None, 1], [1, 1, 3], [4, 2, 4], [5, 3, None], [5, 1, 4]]
df = pd.DataFrame(list2d)
df
```

```python
def k_means(arr2d, b1, b2):
    cls = []
    #
    for arr in arr2d:
        dist1 = 0
        dist2 = 0
        for val, val1,val2 in zip(arr, b1, b2):
            if val == None:
                continue
            dist1 += (val-val1) ** 2
            dist2 += (val-val2) ** 2
        cls.append(dist2 > dist1)
    cls = [0 if c else 1 for c in cls]
    
    #
    b = np.array([[0, 0, 0], [0, 0, 0]])
    cnt = np.array([[0, 0, 0], [0, 0, 0]])
    
    for arr, c in zip(arr2d, cls):
        for i, val in enumerate(arr):
            if val == None:
                continue
            b[c][i] += val
            cnt[c][i] += 1
    
#     print(b)
#     print(cnt)
    b = b/cnt
    b1 = b[0]
    b2 = b[1]
    
    return b1, b2, cls
```

```python
b1=[3, 0, 3]
b2=[3, 3, 5]
cls = []

while(1):
    print(np.array(cls)+1)
    cls_prev = cls
    b1, b2, cls = k_means(list2d, b1, b2)
    if cls == cls_prev:
        break

print(np.array(cls)+1)
```
