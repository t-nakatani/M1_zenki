```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/exercise_6.csv', index_col=0)
x1 = np.array(df['x1'])
x2 = np.ones(x1.shape)
X = np.array([x1, x2]).T
y = np.array(df['y'])
e2 = np.array(df['e^2'])
# w_old = np.array([0.4545, 2.0])
w = 2  / (e2+1)**2
W = np.diagflat(w, k=0)

a = np.linalg.inv(X.T @ W @ X) @ X.T @ W @ y
print(a)
plt.figure()
plt.scatter(np.arange(1, 11), y)
plt.plot(np.arange(1, 11), X @ a.T)
plt.show()
```
