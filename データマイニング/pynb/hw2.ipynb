```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/exercise_2.csv', index_col=0)
df
```

```python
X = np.array(df.loc[:, 'x1':'x3'])
```

```python
XTX = X.T@X
inv_XTX = np.linalg.inv(XTX)
inv_XTX
```

```python
a = inv_XTX@X.T@np.array(df['y'])
print('x1の係数:{}, x2の係数:{}, x3の係数{}'.format(a[0], a[1], a[2]))
```
